```python
# Startup code for the SVD class ...

class SVDReps:
    """
    The Poly2D class builds and plots 2D polynomials.
    
    Attributes:
    -----------
    A:             this is the input matrix.
    U, Sigma, Vh:  this is the SVD decomposition for A.
    rank:          this is the effective rank for A.
    A_plus:        this is the pseudo-inverse of A.
    
    ColumnSpace:   Matrix to project on Column space.
    NullSpace:     Matrix to project on the Null space.
    RowSpace:      Matrix to project on the Row space.
    LeftNullSpace: Matrix to project on left null space.
    ... Describe any other attributes as needed
    
    Methods:
    --------
    Constructor requires A and generates projections.
    projectColumnSpace(), projectNullSpace(), projectRowSpace(), projectLeftNullSpace():
        project b or x to the different spaces of A.
    ... ADD more methods as needed.
    """
    
    # Constructor
    def __init__(self, A, threshold=1):
        self.A = A  # Create an instance variable
        self.U, self.Sigma, self.Vh = np.linalg.svd(A)
        self.threshold = threshold
        # Use the sigmas to extract the rank:
        # Write a for-loop to determine where sigma1/sigma_rank is HIGH.
        # Try 100 but you must also experiment with at-least one more threshold.
        self.rank = 0
        for sig  in self.Sigma:
            if self.Sigma[0]/sig < self.threshold:
                self.rank += 1
        
        # Once you have the rank, you can prepare all of the projections:
        # REVIEW Stanford lecture on how to extract elements of a matrix.
        # REFER to slide 12 of NLA.pdf for what you need to extract and multiply.
        # Use .dot to multiply matrices. Use .T to transpose matrices.
        self.ColumnSpace   = np.dot(self.U[:, :self.rank], np.conj((self.U[:, :self.rank]).T))
        self.NullSpace     = np.dot(self.Vh.T[:, self.rank:], np.conj((self.Vh.T[:, self.rank:]).T)) # DEFINE 
        self.RowSpace      = np.dot(self.Vh.T[:, :self.rank], np.conj((self.Vh.T[:, :self.rank]).T)) # DEFINE 
        self.LeftNullSpace = np.dot(self.U[:, self.rank:], np.conj((self.U[:, self.rank:]).T)) # DEFINE 
        
        
        # Construct the pseudo-inverse as discussed in class:
        self.S_plus = np.zeros(self.A.T.shape)
        for i in range(self.rank):
            self.S_plus[i, i] = 1/self.Sigma[i]
            
        self.A_plus = self.Vh.T @ self.S_plus @ self.U.T # DEFINE using U, Sigma, Vh, .T, and .dot()
    
     # Instance methods
    def plot(self, A, space):
        plt.figure(figsize=(10, 10))
        origin = np.array([[0, 0], [0, 0]])
        for i in range (len(A.T)):
            plt.quiver(*origin, A[0, i], A[1, i], angles='xy', scale_units='xy', scale=1.)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.xlim([-1.1,1.1])
        plt.ylim([-1.1, 1.1])
        plt.grid(1)
        plt.title(space)
        plt.show()
    
    # Instance methods
    def projectColumnSpace(self, b):
        """ Projects b against the Column Space
        """
        proj = np.dot(self.ColumnSpace, b)
        return proj
    
    
    def visColumnSpace(self, n_rows, n_cols):
        """ Displays the Column Space of A using the SVD decomp.
            If this is a 1D example, set n_rows=n_cols=1.
            Else, the function should display a collection of 2D images.
        """
        # Write a for-loop that displays U_1, ..., U_r from slide 12 of NLA.pdf
        for i in range(self.rank):
            if (n_rows == 1) and (n_cols == 1):
                print(self.U[:,i])
            else:
                # Write the code for imshow() here.
                self.plot(self.ColumnSpace, 'Column Space') 
                return 
        
    # MODIFY the rest of them to be like projectColumnSpace()
    def projectNullSpace(self, x):
        proj = np.dot(self.NullSpace, x)
        return proj
    
    def visNullSpace(self, n_rows, n_cols):
        for i in range(self.rank):
            if (n_rows == 1) and (n_cols == 1):
                print(self.Vh[:,-i])
            else:
                # Write the code for imshow() here.
                self.plot(self.NullSpace, 'Null Space') 
                return 
    
    def projectRowSpace(self, x):
        proj = np.dot(self.RowSpace, x)
        return proj
    
    def visRowSpace(self, n_rows, n_cols):
        for i in range(self.rank):
            if (n_rows == 1) and (n_cols == 1):
                print(self.Vh[:,i])
            else:
                # Write the code for imshow() here.
                self.plot(self.RowSpace, 'Row Space') 
                return 
     
        
    def projectLeftNullSpace(self, b):
        proj = np.dot(self.LeftNullSpace, b)
        return proj
    
    def visLeftNullSpace(self, b, n_rows, n_cols):
        for i in range(self.rank):
            if (n_rows == 1) and (n_cols == 1):
                print(self.U[:,-i])
            else:
                # Write the code for imshow() here.
                self.plot(self.LeftNullSpace, 'Left Null Space') 
                return
    
    
    def LowVecs(self):
        """ Projects b against the Column Space
        """        
        # Extract them from the SVD decomposition of A.
        sigma_r = self.Sigma[:self.rank] # it is an 1D array
        u_r = self.U[:, :self.rank]
        v_r = self.Vh[:, :self.rank]
        return sigma_r, u_r, v_r
    
    # IMPLEMENT the rest of the methods below.
    # ...
    
    def LeastSQ(self, b):
        """ Computes x that gives the least-squares solution to Ax=b
        """
        # Remember that: A = U*Sigma*V^T
        # Thus, A^+ = V*Sigma^{-1}*U^T 
        #  where Sigma^{-1} has 1/sigma_i along the diagonal and zero
        #  everywhere else.
        # ... Use x = A^+ b
#         S = np.zeros(self.A.shape)
#         for i in range(len(self.Sigma)):
#             S[i, i] = self.Sigma[i]
#         x = self.Vh.T @ np.linalg.pinv(S) @ self.U.T @ b
        x = self.A_plus @ b
        return x
    
    def RCNs(self, x, b):
        """ Computes the RHS of the four condition numbers and returns them.
        """
        # ... See lecture notes for definitions of the quantities
        # The norm is given by numpy.linalg.norm
        # The condition number is given by numpy.linalg.cond
        # With these two, you just need to add the inverse trig functions ...
        Ax = self.A@x
        cond =  np.linalg.norm(Ax) / np.linalg.norm(b)
        if cond > 1:
            cond = 1
        theta = np.arccos(cond)
        n = np.linalg.norm(self.A) * np.linalg.norm(x) / np.linalg.norm(Ax)
        k = self.Sigma[0]/self.Sigma[self.rank-1]
        s1 = 1 / np.cos(theta)
        s2 = k / (n * np.cos(theta))
        s3 = k / np.cos(theta)
        s4 = k + ((k**2) * np.tan(theta) / n)
        return s1, s2, s3, s4
    
    def reshape(self, NumComps):
        Uc = self.U[:, :NumComps]
        Vhc = self.Vh[:NumComps, :]
        Sc = np.diag(self.Sigma[:NumComps])
        return Uc, Sc, Vhc
    
    def compressA(self, NumComps):
        """ Compresses A given the NumComps to keep.
        """
        # Reshape the matrices in A=U Sigma Vh so that 
        #    only NumComps are kept in each matrix.
        # Then use them to calculate W1, W2 as given in the homework
        #    announcement.
        m, n = self.A.shape
        Uc, Sc, Vhc = self.reshape(NumComps)
        CompressionRatio = n * m / (NumComps * (n + m)) # See lecture notes how
#         W, W1, W2        = self.U @ Sc @ self.Vh,  Sc @ self.Vh, self.U # Calculate these three matrices.
        W, W1, W2        = Uc @ Sc @ Vhc,  Sc @ Vhc, Uc # Calculate these three matrices.
        return W, W1, W2, CompressionRatio
    
    def getNumStableComps(self, CondBound):
        """ Gets the number of components to so that the new Condition number < CondBound
        """
        # Write a simple for-loops that goes through the sigma_i.
        # Find NumComps such that  sigma_1/sigma_NumComps < CondBound using a for-loop.
        # Return NumComps
        NumComps = 0 # Determine as given above.
        sig_o = self.Sigma[0]
        for sigma_NumComps in self.Sigma:
            if sig_o / sigma_NumComps < CondBound:
                NumComps += 1
        return NumComps
        
    def getCompsBounds(self, NumComps):
        """ Returns totalFrobenius, FrobeniusError, ConditionNumber when using A_k 
            to approximate A where k=NumComps denotes the number of components that are kept.      
        """
        # Refer to the slides for how to compute
        #     totalForbenius using sigma_i, FrobeniousError.
        # Refer to RCNs on how to get the ConditionNumber
        Uc, Sc, Vhc = self.reshape(NumComps)
        totalFrobenius = np.sqrt((self.Sigma[:NumComps]**2).sum())
        FrobeniusError = np.sqrt((self.Sigma[NumComps:]**2).sum())
        ConditionNumber = self.Sigma[0] / self.Sigma[NumComps-1]
        return totalFrobenius, FrobeniusError, ConditionNumber

# Example of A, b from the first lecture:
# Make sure that the projections ARE EXACTLY the same as given in the lecture.
# Note that the U, Sigma, V^T may be slightly different. 
A = np.array([[1, 3], [2, 6]])
b = np.array([[4], [8]])

# Construct an instance of the SVDReps class
Asvd = SVDReps(A) 

# Project b one the column space of A
b_cs = Asvd.projectColumnSpace(b)
```

```python
def show_slices(slices, cmap="gray"):
    # This function comes from the nibabel
    # tutorial --> https://nipy.org/nibabel/coordinate_systems.html#introducing-someone
    
    fig, axes = plt.subplots(1, len(slices))

    for i, slice in enumerate(slices):
        im = axes[i].imshow(slice.T, cmap=cmap, origin="lower")
        
    plt.show()
    
def contour_slices(slices, cmap="gray", total_levels=6):
    # This function includes also plotting contour plot
    # link --> https://www.python-course.eu/matplotlib_contour_plot.php
    # link --> https://github.com/silx-kit/silx/issues/2242
    
    fig, axes = plt.subplots(1, len(slices), figsize=(5,5))
    contour_levels = total_levels
    titles = ['sagital', 'coronal', 'axial']
    
    for i, slice in enumerate(slices):
        row, col = np.shape(slice)
        y = np.arange(0, row)
        x = np.arange(0, col)
        xx, yy = np.meshgrid(x, y)

        
        zzmin, zzmax = np.min(slice), np.max(slice)
        levels = np.linspace(zzmin, zzmax, contour_levels)

        #Display image with contour plot
        im = axes[i].imshow(slice.T, cmap=cmap, origin="lower")
        contour = axes[i].contour(yy, xx, slice, levels)
        axes[i].clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
        c = ('#ff0000', '#ffff00', '#0000FF', '0.6', 'c', 'm')
        contour_filled = axes[i].contourf(yy, xx, slice, colors=c)

        
    plt.show()
```
